```python
import requests
from bs4 import BeautifulSoup
from bs4 import NavigableString
```

```python
url = requests.get("https://www.latlong.net/category/hospitals-102-26.html").text
```

```python
soup = BeautifulSoup(url,"lxml")
```

```python
soup
```

```python
table = soup.find("table")
```

```python
table
```

```python
links = []
links.extend(table.find_all("a",href = True))
links_of_hospitals = ["https://www.latlong.net"+links[i]["href"] for i in range(len(links))]
```

```python
links_of_hospitals[:5]
```

```python
tr_list = []
tr_list.extend(table.find_all("tr"))
```

```python
Name_of_Hospital = []
Latitude = []
Longitude = []
```

```python
flag = 0
for i in range(1,len(tr_list)):
    td_list = tr_list[i].find_all("td")
    for td in td_list:
        td_kids = list(td.children)
        
        #Name_of_Hospital.append(td_kids[0].get_text())
        #Latitude.append(float(td_kids[1]))
        #Longitude.append(float(td_kids[2]))
        
        for val in td_kids:
           
            if isinstance(val, NavigableString):
                if flag == 0:
                    Latitude.append(float(val))
                    flag+=1
                else:
                    Longitude.append(float(val))
            else: 
                Name_of_Hospital.append((val.get_text()))
                flag = 0
```

```python
Name_of_Hospital[:5]
```

```python
Latitude[:5]
```

```python
Longitude[:5]
```

```python
import pandas as pd
```

```python
Hospitals = pd.DataFrame(list(zip(Name_of_Hospital, Latitude, Longitude, links_of_hospitals)),
                        columns=["Name","Latitude","Longitude","url"])
```

```python
Hospitals.head()
```

```python
import folium
```

```python
loc = Hospitals[ ['Latitude', 'Longitude'] ]
```

```python
loca_list = loc.values.tolist()
```

```python
lat = 21
long = 78
```

```python
m =folium.Map(location=[lat,long], zoom_start=5)
```

```python
k =0
for loca in loca_list:
    folium.Marker(location=loca,
                  tooltip=Hospitals.iloc[k][0], 
                  popup = folium.Popup('<a href={}   target="_blank">' "Click to now about this Hospital" '</a>'.format(Hospitals.iloc[k][3])),
                  draggable = False,
                  icon=folium.Icon(color='red', icon='hospital-o', prefix='fa', icon_color = "white")).add_to(m)
    
    k+=1
```

```python
m
```

```python
m.save('hospitalsofindia.html')
```
